#### features (Columns) in the dataset and meaning:

 * age - age in years 
 * sex - (1 = male; 0 = female)
 * cp - Chest pain: type(0: no pain, 1: typical angina, 2: atypical angina, 3: non-anginal pain, 4: asymptomatic)
 * trestbps - resting blood pressure( in mm Hg on admission to the hospital)
 * chol - serum cholestoral in mg/dl
 * fbs - (Fasting blood sugar >120mg/dl) (1 = true, 0 = false)
 * restecg - resting electrocardiographic results.
 * thalach - maximum heart rate achieved.
 * exang - exercise induced angina ( 1 = yes, 0 = no)
 * Oldpeak - ST depression induced by exercise relative to rest
 * Slope - the slope of the peak exercise ST segment
 * ca - number of major vessels (0-3) colored by flourospy
 * thal - thalassemai (1 = normal; 2 = fixed defect; 3 = reversable defect)
 * target - have disease or not ( 1 = yes, 0 = no).

```python
def Shorten_categories(categories):
    categorical_grp = {}
    for i in range (len(categories)):
        if categories.values[i] >= 50:
            categorical_grp[categories.index[i]] = categories.index[i]
        else:
            categorical_grp[categories.index[i]] = 1
    return categorical_grp
```

```python
Thal_grp = Shorten_categories(df.thalassemai.value_counts())
df['thalassemai'] = df['thalassemai'].map(Thal_grp)
```

```python
# Data Visualization for age bracket

def age_bracket(age):
    if age <= 35:
        return "Youth(<=35)"
    elif age <= 55:
        return "Adult(<=55)"
    elif age <= 65:
        return "Old Adult(<=65)"
    else:
        return "Elderly(>65)"
    
df["age_bracket"] = df["age"].apply(age_bracket)

# Investigating the age group of patients
plt.figure(figsize =(10,5))
ax = sns.countplot(x = "age_bracket", data=df)
plt.xlabel("age_bracket")
plt.ylabel("count of age group")
plt.title("Total number of patients")
ax.bar_label(container=ax.containers[0], label="Count of age group");   
```

### FEATURE ENGINEERING / DATA PRE-PROCESSING

#### To prepare the data for the machine learning model, all the features must be numerical (integer or float)
#### All objects or categorical data must be handled (converted to integer or removed from the data.)


```python
df.head(2)
```

#### From the table above, age_bracket, gender, cp_cat, ht_d and label are all categorical data that were constructed, hence

#### we are going to drop them. 

#### More also, since TARGET is what we are predicting, we are going to drop TARGET ( We dont intend to give the machine

#### answers to what it should predict)

```python
#  Create a copy of the data (Exclude target/ label alongside other columns that were created)

df1 = df[['age', 'sex', 'Chest_pain_type', 'resting_blood_pressure',
       'cholesterol', 'Fasting_blood_sugar', 'restecg',
       'max_heart_rate_achieved', 'exercise_induced_angina ', 'st_depression',
       'st_slope', 'num_of_major_vessels', 'thalassemai']]



label = df[["target"]]
```

```python
df1
```

```python
label.head()
```

```python
# Check the data types ( the new copy made) to be sure all columns are numerical values.

df1.dtypes
```

### Dealing with the features with OUTLIERS

#### -handling the outliers through NORMALIZATION (scaling the values to be between 0 and 1)

#### - After normalizing ( Applying the scaler to columns with outliers), we drop the original columns, moving foward with the new scaled columns

```python
scaler = MinMaxScaler()

df1['Scaled_RBP'] = scaler.fit_transform (df1['resting_blood_pressure'].values.reshape(-1, 1))
df1['Scaled_Chol'] = scaler.fit_transform (df1['cholesterol'].values.reshape(-1, 1))
df1['Scaled_thal'] = scaler.fit_transform(df1['thalassemai'].values.reshape(-1, 1))
df1['Scaled_max_heart_rate'] = scaler.fit_transform (df1['max_heart_rate_achieved'].values.reshape(-1, 1))
    
# Drop the original columns

df1.drop(['resting_blood_pressure', 'cholesterol', 'thalassemai', 'max_heart_rate_achieved'], axis = 1, inplace=True)
```

```python
df1.head()
```

### MACHINE LEARNING

#### - Split the dataset into training and testing sets 

#### (X = questions, y =  Answers)

#### test_size = 0.2( Using 20% of the data to TEST the model, The remaining 80% to train the model)

#### random_state = 42 (instructing the system to select the rows at random, NOT picking 80% of the data serially)

```python
X_train, X_test, y_train, y_test = train_test_split(df1, label, test_size = 0.2, random_state = 42)
```

```python
#To view X_train data

X_train
```

```python
# To view the y_train data (the actual disease status)
y_train
```

### MODEL BUILDING

```python
# USING "LOGISTIC REGRESSION" ALGORITHM

logreg = LogisticRegression()


# Use the logistic regression to train your data

logreg.fit(X_train, y_train)

# Use the trained "logreg" to predict your test data

ly_pred = logreg.predict(X_test)

# Apply the all the evaluation metrices 

print("Logistic Regression")
print("Accuracy:", accuracy_score(y_test, ly_pred))
print("Precision:", precision_score(y_test, ly_pred))
print("Recall:", recall_score(y_test, ly_pred))
print("F1-score:", f1_score(y_test, ly_pred))
print("AUC-ROC:", roc_auc_score(y_test, ly_pred))
```

### Deploying multiple machine learning algorithm at once

Applying 8 machine learning algorithm to the dataset

```python
# Create a list of the 8 algorithms. aAd the name of the list here is CLASSIFIERS


Classifiers = [[XGBClassifier(), 'XGB Classifier'],
              [RandomForestClassifier(), 'Random Forest'],
              [KNeighborsClassifier(),'K-Nearest Neighbors'],
              [SGDClassifier(), 'SGD Classifier'],
              [GaussianNB(), 'Naive Bayes'],
              [SVC(), 'SVC'],
              [DecisionTreeClassifier(random_state = 42), 'Decision Tree'],
              [LogisticRegression(), 'Logistic Regression']]
```

```python
# Run the classifier

Classifiers
```

```python
# Create a dictionary of the evaluation metrices.

acc_list = {}
precision_list = {}
recall_list = {}
roc_list = {}
f1_list = {}



for classifier in Classifiers:
    model = classifier [0]
    model.fit(X_train, y_train)
    model_name = classifier [1]
    
    pred = model.predict(X_test)
    
    a_score = accuracy_score(y_test, pred)
    p_score = precision_score(y_test, pred)
    r_score = recall_score(y_test, pred)
    roc_score = roc_auc_score(y_test, pred)
    f_score = f1_score(y_test, pred)
    
# Converting the scores into a percentatge

    acc_list[model_name] = ([str(round(a_score*100, 2)) + '%'])
    precision_list[model_name] = ([str(round(p_score*100, 2)) + '%'])
    recall_list[model_name] = ([str(round(r_score*100, 2)) + '%'])
    roc_list[model_name] = ([str(round(roc_score*100, 2)) + '%'])
    f1_list[model_name]= ([str(round(f_score*100, 2)) + '%'])
    
    if model_name != classifier[-1][1]:
        print('')
```
